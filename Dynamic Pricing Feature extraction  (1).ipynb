```python
correlation_matrix = data.corr()
threshold = 0.7
correlation_features = set()
correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
            correlated_data = data.drop(correlated_features, axis=1)

print(correlated_data.head())
```

```python
selected_features = ['Number_of_Riders' , "Number_of_Drivers", 'Location_Category','Customer_Loyalty_Status', 
                     'Number_of_Past_Rides','Average_Ratings','Time_of_Booking', 'Vehicle_Type','Historical_Cost_of_Ride']
domain_based_feature = data[selected_features]
print(domain_based_feature.head())
```

```python
data = domain_based_feature.copy()

#the balance between supply and demand in the context of dynamic pricing.

data['Number_of_Riders_and_driver'] = data['Number_of_Riders']/ data['Number_of_Riders'] 


data['Cost_Per_Past_Ride'] = data['Historical_Cost_of_Ride']/ data['Number_of_Past_Rides']
```

```python
#handle text and categorical data 
# Perform one-hot encoding for categorical columns
data = pd.get_dummies(data, columns=['Location_Category', 'Vehicle_Type'])
```
